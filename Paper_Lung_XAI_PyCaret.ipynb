```python
import pandas as pd
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.metrics import (
    roc_curve, auc, accuracy_score, f1_score, precision_score,
    recall_score, confusion_matrix
)
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.preprocessing import StandardScaler
from catboost import CatBoostClassifier
from xgboost import XGBClassifier
import sklearn
from sklearn.model_selection import StratifiedKFold, cross_val_predict

# 1. Load dataset
target_column = 'ajcc.stage'
# 2. Define feature sets
feature_sets = {
    'gs1': ['BTG1', 'JUND', 'IER3', 'ZNF331', 'PSAP'],
    'gs2': ['ATP6V0D1', 'ATP6V1B1', 'DRAM2', 'GPSM1', 'LRRK2', 'MAPK3', 'PINK1', 'RRAGB', 'AKT2', 'CIDEC', 'HTRA2', 'PTGIS', 'STK24', 'BAG4', 'CASP4', 'TNFRSF12A', 'TNFRSF8', 'TRADD'],
    'gs3': ['CHMP4C', 'IL1B', 'JAK1', 'PYGB', 'TNFRSF10B'],
    'gs4': ['KLF6', 'SGMS2'],
    'gs5': [ 'MMP20', 'DYNAP', 'CASP14', 'FAM71E2', 'OPN4', 'CGB5', 'DIRC1', 'EQTN', 'FAM205A', 'RTL3'],
    'gs6': ['FGF4', 'FGL1', 'LIM2', 'NPY', 'F13A1', 'CDH12', 'CD1E', 'OTX2', 'ADRA1D', 'SAMD9L', 'ZFP42', 'GAGE2A', 'KLRC2'],
    'gs7': ['ARRB1', 'DGKA', 'FGG', 'EHD1', 'MMRN1', 'DOCK9'],
    'gs8': ['A2M', 'CHEK2', 'ELN', 'FOS', 'PLAU'],
    'gs9': ['SNAI1', 'SMAD7', 'BMP2', 'RGS3'],
    'gs10': ['MBNL2', 'ATP5MG', 'RAMAC', 'MDH1', 'STXBP1', 'ZFP36L2', 'APBB2', 'PDLIM3', 'MYADM', 'PHF5A', 'SLC26A9', 'MAGI2-AS3', 'LINC01089'],
    'gs11': ['CD4', 'NLRP3', 'NT5E', 'TLR4'],
    'gselc38': ['ALDH3B1', 'ALG1L', 'C7', 'CAV1', 'CENPI', 'GKN2', 'GPRC5A',
       'H2AC14', 'LINC00887', 'MMP20', 'PCAT7', 'TNS4', 'KRT17P1',
       'GSDMC', 'HMGB3P6', 'CSTL1', 'GCGR', 'BMP7', 'TFAP2A', 'PLXNB3',
       'CDT1', 'COL9A3', 'TMPRSS4', 'LINC02830', 'GNG4', 'PPFIA4',
       'CACNA1I', 'PLAAT1', 'ONECUT3', 'IMPDH1P8', 'RCOR2', 'FTCD',
       'PPP1R27', 'GINS2', 'ESCO2', 'MYOSLID', 'PDE10A', 'DUXAP9'],
    'gselc20': ['COL9A3', 'H2AC14', 'LINC02830', 'CSTL1', 'MYOSLID',
                'PLXNB3', 'DUXAP9', 'GKN2', 'MMP20', 'C7', 'ESCO2',
                'IMPDH1P8', 'PCAT7', 'ONECUT3', 'PLAAT1', 'ALG1L',
                'GINS2', 'LINC00887', 'ALDH3B1', 'TMPRSS4']
}

# 3. Define models
models = {
    "CAT": CatBoostClassifier(iterations=300, learning_rate=0.1, random_seed=42),
    "XGBoost": XGBClassifier(random_state=42),
    "SVMlinear": sklearn.svm.SVC(kernel="linear", probability=True, random_state=42),
    "SVMradial": sklearn.svm.SVC(kernel="rbf", probability=True, random_state=42),
    "Random Forest": RandomForestClassifier(random_state=42),
    "LR": sklearn.linear_model.LogisticRegression(solver="newton-cg", random_state=42, max_iter=1000),
}

# 4. Train/test split

X_full = df.drop(columns=[target_column])
y = df[target_column]


from imblearn.over_sampling import SMOTE
smote = SMOTE()
X_res1, y_res1 = smote.fit_resample(X_full, y)

cv = StratifiedKFold(n_splits=5)

results = []

# 5. Loop through each combination
for fs_name, features in feature_sets.items():
    X = X_res1[features]

    for model_name, model in models.items():
        y_pred_prob = cross_val_predict(model, X, y_res1, cv=cv, method="predict_proba")[:, 1]
        y_pred = cross_val_predict(model, X, y_res1, cv=cv, method="predict")

        # Calculate  metrics
        fpr, tpr, _ = roc_curve(y_res1, y_pred_prob)
        roc_auc = auc(fpr, tpr)

        accuracy = accuracy_score(y_res1, y_pred)
        f1 = f1_score(y_res1, y_pred)
        precision = precision_score(y_res1, y_pred)
        recall = recall_score(y_res1, y_pred)

        # Sensibility e specificity
        cm = confusion_matrix(y_res1, y_pred)
        tn, fp, fn, tp = cm.ravel()
        sensibility = tp / (tp + fn)
        specificity = tn / (tn + fp)

        # Adicionar resultados na lista
        results.append({
            "Feature Set": fs_name,
            "Model": model_name,
            "Accuracy": accuracy,
            "F1 Score": f1,
            "Precision": precision,
            "Recall": recall,
            "Sensibility": sensibility,
            "Specificity": specificity,
            "AUC": roc_auc
        })
```
